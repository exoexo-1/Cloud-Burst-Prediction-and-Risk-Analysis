```python
print(df['Humidity3pm'].isnull().sum())
print(df['Humidity9am'].isnull().sum())
```

```python
print(df['Humidity3pm'].median() - df['Humidity9am'].median())
print(df['Humidity3pm'].max())
print(df['Humidity3pm'].min())
```

```python
df.fillna({'Humidity9am': df['Humidity9am'].median()}, inplace=True)
df.fillna({'Humidity3pm': df['Humidity3pm'].median()}, inplace=True)

print(df['Humidity3pm'].isnull().sum())
print(df['Humidity9am'].isnull().sum())

print(df.isnull().sum())
```

```python
df.fillna({'Pressure9am': df['Pressure9am'].median()}, inplace=True)
df.fillna({'Pressure3pm': df['Pressure3pm'].median()}, inplace=True)

print(df.isnull().sum())
```

```python
print(df['CloudBurst Today'].value_counts())
```

```python
df.fillna({'WindGustSpeed': df['WindGustSpeed'].median()}, inplace=True)
print(df.isnull().sum())
```

```python
df.fillna({'MinTemp': df['MinTemp'].median()}, inplace=True)
df.fillna({'MaxTemp': df['MaxTemp'].median()}, inplace=True)
df.fillna({'Temp9am': df['Temp9am'].median()}, inplace=True)
df.fillna({'Temp3pm': df['Temp3pm'].median()}, inplace=True)
df.fillna({'WindSpeed9am': df['WindSpeed9am'].median()}, inplace=True)
df.fillna({'WindSpeed3pm': df['WindSpeed3pm'].median()}, inplace=True)
```

```python
print(df.isnull().sum())
print(df.shape)
```

```python
# Fill missing wind directions with mode (most frequent value)
df.fillna({'WindDir9am': df['WindDir9am'].mode()[0]}, inplace=True)
df.fillna({'WindDir3pm': df['WindDir3pm'].mode()[0]}, inplace=True)
df.fillna({'WindGustDir': df['WindGustDir'].mode()[0]}, inplace=True)
print(df.isnull().sum())
```

```python
df[["MinTemp", "MaxTemp", "Temp9am", "Temp3pm", "CloudBurstTodayBinary"]].corr()
```

```python
# Convert 0-10 scale to 0-100%
df['Cloud9am'] = df['Cloud9am'] * 10
df['Cloud3pm'] = df['Cloud3pm'] * 10
```

```python
#After predictive impudation of Evaporation
new_df = pd.read_csv("clean_cloudburst_data.csv")
```

```python
print(new_df.isnull().sum())
print(new_df.shape)
```

```python
print(new_df['WindDir3pm'].value_counts())
```

```python
wind_dir_mapping = {
    'N':0, 'NNE':22.5, 'NE':45, 'ENE':67.5, 'E':90, 'ESE':112.5, 'SE':135,
    'SSE':157.5, 'S':180, 'SSW':202.5, 'SW':225, 'WSW':247.5, 'W':270,
    'WNW':292.5, 'NW':315, 'NNW':337.5
}
new_df['WindDir9am'] = new_df['WindDir9am'].map(wind_dir_mapping)
new_df['WindDir3pm'] = new_df['WindDir3pm'].map(wind_dir_mapping)
```

```python
new_df['WindDir9am_sin'] = np.sin(np.deg2rad(new_df['WindDir9am']))
new_df['WindDir9am_cos'] = np.cos(np.deg2rad(new_df['WindDir9am']))
new_df['WindDir3pm_sin'] = np.sin(np.deg2rad(new_df['WindDir3pm']))
new_df['WindDir3pm_cos'] = np.cos(np.deg2rad(new_df['WindDir3pm']))

# Drop the original columns
new_df.drop(columns=['WindDir9am','WindDir3pm'], inplace=True)
```
